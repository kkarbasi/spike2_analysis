```python
# csv list of david's neuron filenames with the size of the spike trains

neurons_path = []
with open('../data/neurons_david_data_with_spike_train.csv', 'rb') as csvfile:
    freader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in freader:
        neurons_path = neurons_path + row

from __future__ import print_function
with open('all_davids_neuros_num_spikes_in_spike_trains.csv', 'w') as f:
    for npath in neurons_path:
        neo_reader = Spike2IO(filename=npath)
        neo_data = neo_reader.read()
        data_block = neo_data[0]
        seg = data_block.segments[0]
        to_print = npath
        for st in seg.spiketrains:
            to_print = to_print + ',' + str(st.size)
        print(to_print, file=f)
```

```python
# Check which files from david's list of used neurons have spike trains in them
from __future__ import print_function
with open('all_davids_neuros.out', 'w') as f:
    for npath in neurons_path:
        for root, dirnames, filenames in os.walk(npath):
            for filename in filenames:
                if filename.endswith('smr'):
                    smr_fn = os.path.join(root, filename)
                    try:
                        neo_reader = Spike2IO(filename=smr_fn)
                        neo_data = neo_reader.read()
                        data_block = neo_data[0]
                        seg = data_block.segments[0]
                        if seg.spiketrains: 
                            print('{}'.format(smr_fn), file=f)
#                             print('>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>', len(seg.spiketrains), file=f)
                    except AssertionError as e:
                        print(repr(e))
#         print("--------------------------------", file=f)

```
